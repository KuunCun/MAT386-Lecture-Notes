## Motivation for Logistic Regression

For binary classification, we will focus on models of the form:

$$ 
f_{\theta}(\textbf{x}_i) = \sigma(\boldsymbol{\theta}^\top \textbf{x}_i) = \frac{1}{1 + \exp(-\boldsymbol{\theta}^\top \textbf{x}_i)}, 
$$

where $i=1,\ldots, n$. 

After training the model $f_{\theta}(\textbf{x}_i)= \sigma(\boldsymbol{\theta}^\top \textbf{x}_i) = \frac{1}{1 + \exp(-\boldsymbol{\theta}^\top \textbf{x}_i)}$ on the training dataset $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$, here, our aim is to **assign a new observation** $\mathbf{x}_{n+1}$ into one of the classes $\{0,1\}$. 

When the `linear predictor` $\boldsymbol{\theta}^\top \textbf{x}_i=z_i$, the expression above is equal to the following:

$$
\sigma(z_i) = \frac{1}{1 + \exp(-z_i)},
$$

which is known as the `logistic` function. For that reason, to better understand logistic regression, let's take a look at the logistic function first.

## Details on Logistic function

More specifically, 
the logistic function $\sigma:\mathbb {R} \rightarrow (0,1)$ is defined as follows:


\begin{equation}
\sigma(z) := \frac{\exp(z)}{1+\exp(z)}=\frac{1}{1+\exp(-z)}.
\nonumber
\end{equation}

The logistic function $\sigma$ takes any real input $z$ and "squeezes" inputs from the real line into $(0,1)$.

The logistic function is also a `sigmoid function`, in other words, an `S-shaped function` as can be seen below:

```python
#create z vector within a specified interval
import numpy as np

def sigmoid(z):
    return 1/(1+np.exp(-z))

z = np.linspace(start=-10, stop=10, num=50)
```

```python
#plot z vs sigmoid(z)
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [16, 6]

plt.plot(z, sigmoid(z), c="r")
plt.axvline(x = 0, color = 'black', label = 'axvline - full height', linestyle='dashed')
plt.xlabel("z")
plt.ylabel("σ(z)")
plt.show();
```

## Logistic Function Properties

The logistic function $\sigma(z)$ function tends to 1 as $z \to \infty$ such that: 

$$
\lim_{z\to\infty} \sigma(z)=\lim_{z\to\infty} \frac{1}{1+\exp(-z)}=1
$$ 

and it tends to 0 as $z \to -\infty$ such that: 

$$
\lim_{z\to -\infty} \sigma(z)=\lim_{z\to -\infty} \frac{1}{1+\exp(-z)}=0.
$$

The derivative of $\sigma(z)$ has a simple form:

$$
\frac{d\sigma(z)}{dz} = \frac{d}{dz} \bigg(\frac{1}{1+\exp(-z)}\bigg) = \frac{\exp(-z)}{\big(1+\exp(-z)\big)^2}= \bigg(\frac{1}{1+\exp(-z)}\bigg)\bigg(\frac{\exp(-z)}{1+\exp(-z)}\bigg)=\sigma(z)(1-\sigma(z)).
$$

**How can we integrate the logistic function to binary classifacation problems**?

## Logistic Regression

**Given a training dataset** $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$ of size $n$, for binary classification problems with $y_i \in \{0,1\}$, the `logistic regression` builds a functional form between $Pr(y_{i}=1 \vert \mathbf{x}_{i})$ 
and the `logistic function` of the `linear predictor`, $z_i=\boldsymbol{\theta}^\top \textbf{x}_i=\theta_0+\sum_{j=1}^{d}\theta_jx_{ij}$, which is the linear combination of a set of $d$ features and 
their corresponding coefficients, as follows:


\begin{eqnarray}
\begin{split}
Pr(y_{i}=1\vert \mathbf{x}_{i}) &= \sigma(\boldsymbol{\theta}^\top \textbf{x}_i)=\frac{\exp(\boldsymbol{\theta}^\top \textbf{x}_i)}{1+\exp(\boldsymbol{\theta}^\top \textbf{x}_i)}=\frac{1}{1+\exp(-\boldsymbol{\theta}^\top \textbf{x}_i)},
\nonumber
\end{split} 
\end{eqnarray}

where 


$$
Pr(y_{i}=0\vert \mathbf{x}_{i}) = 1-Pr(y_{i}=1\vert \mathbf{x}_{i})=1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)=\frac{1}{1+\exp(\boldsymbol{\theta}^\top \textbf{x}_i)},
$$

where $i=1,\ldots,n$.

For example, assume $X_1 = $ hours of study and $Pr(Y = 1 |x_1)$ probability passing an exam depending on the values of $X_1$, are collected from a number of participants. Then the logistic regression models how probability of passing an exam is associated with numbers of studying.

```python
#define Pr(y=1 given x)
import numpy as np

def predict_prob(x1):
       
    z = (1.2*x1) 
    prob = 1/(1+np.exp(-z))
    
    return prob 
```

```python
#investigate how the probability of passing an exam is changing with respect to hours of studying
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [16, 6]

x1 = np.arange(-9, 9, 0.5)  #number of hours take negative values here.

plt.plot(x1, predict_prob(x1), c="r")
plt.xlabel("Hours of Studying")
plt.ylabel("Probability of Passing an Exam")
plt.show();
```

## Odds-Ratio and Logit function

The `odds-ratio (OR)` represents the odds that the occurence of the outcome $y_{i}=1$ given $\mathbf{x}_{i}$ compared to the occurence of the outcome $y_{i}=0$ given $\mathbf{x}_{i}$. The formula of `odds-ratio` is given below:

\begin{eqnarray}
\begin{split}
\text{OR} = \frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert\mathbf{x}_{i})}.
\nonumber
\end{split} 
\end{eqnarray}

An $OR>1$ indicates increased occurence of the outcome $y_{i}=1$.


The `logarithm of odds ratio` defined as `logit function`:


\begin{eqnarray}
\begin{split}
\log(OR)=\log\Bigg(\frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert\mathbf{x}_{i})}\Bigg)= logit(Pr(y_{i}=1\vert \mathbf{x}_{i})).
\nonumber
\end{split} 
\end{eqnarray}

Under logistic regression, `odds-ratio` and `logit function` are defined, respectively, as follows:


\begin{eqnarray}
\begin{split}
\text{OR} &= \frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert\mathbf{x}_{i})}= \exp(\boldsymbol{\theta}^\top \textbf{x}_i) \quad \text{and} \\
\log\Bigg(\frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert\mathbf{x}_{i})}\Bigg)&= logit(Pr(y_{i}=1\vert \mathbf{x}_{i})) =\boldsymbol{\theta}^\top \textbf{x}_i.
\nonumber
\end{split} 
\end{eqnarray}

## HW:
The logistic function $\sigma(z)$ is a `monotonic increasing function` which means that **inverse of the function exists**.

We have `logistic` and `logit` functions defined as follows:


\begin{eqnarray}
\begin{split}
\sigma(z) =\frac{1}{1+\exp(-z)} \quad \text{and} \\  
logit(p) = \log\Big(\frac{p}{1-p}\Big).
\nonumber
\end{split} 
\end{eqnarray}

The **logit function** is the `inverse function` of **logistic function**.


\begin{eqnarray}
\begin{split}
logit\big(\sigma(z)\big) = \log\Big(\frac{\sigma(z)}{1-\sigma(z)}\Big)= \log(\exp(z))=z.
\nonumber
\end{split} 
\end{eqnarray}

## Decision boundary

In **classification**, classes define a **partition of the feature space**, and the **boundary that separate these regions** is called as the `decision boundary`.

The `decision boundary`  between any two class consists of the set of $\mathbf{x}_{i}$ points ($i=1,\ldots,n$) in which `two classes are equally probable` such that:

\begin{equation} 
Pr(y_{i}=1\vert \mathbf{x}_{i}) &= Pr(y_{i}=0\vert \mathbf{x}_{i}),
\nonumber
\end{equation}

which gives us where `odds ratio` equals 1:

\begin{eqnarray}
\begin{split}
\frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert \mathbf{x}_{i})}&=1.
\nonumber
\end{split} 
\end{eqnarray}

Equivalently, where the `logarithm of odds ratio` is 0:

\begin{eqnarray}
\begin{split}
\log\bigg(\frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert \mathbf{x}_{i})}\bigg)=0.
\nonumber
\end{split} 
\end{eqnarray}


Under logistic regression, this is equivalent to:

\begin{eqnarray}
\begin{split}
\log\bigg(\frac{Pr(y_{i}=1\vert \mathbf{x}_{i})}{Pr(y_{i}=0\vert \mathbf{x}_{i})}\bigg)&=\boldsymbol{\theta}^\top \textbf{x}_i=0.
\nonumber
\end{split} 
\end{eqnarray}


The right-hand side of the above equation is the **decision boundary of logistic regression**, $\boldsymbol{\theta}^\top \textbf{x}_i=\theta_0+\sum_{j=1}^{d}\theta_jx_{ij}=0$, which is a **linear function** in $\mathbf{x}_{i}$.

Here, our aim is to learn the decision boundary, which requires learning $\boldsymbol{\theta}$ based on the training dataset $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$.

## How to assign a new observation into one of the classes?

Once we learn $\boldsymbol{\theta}$ from the training dataset $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$, we are able to construct the decision boundary.

Then, we can **assign a new observation** $\mathbf{x}_{n+1}$ into one of the classes $\hat{y}_{n+1} \in \{0,1\}$ **based on** the `estimated decision function` as follows:

\begin{equation}
\hat{y}_{n+1}= 
\begin{cases}
1,  & \text{if} \quad \widehat{\boldsymbol{\theta}}^\top \textbf{x}_{n+1}>0 ,\\
0,  & \text{if} \quad \widehat{\boldsymbol{\theta}}^\top \textbf{x}_{n+1}<0.
\end{cases} \nonumber
\end{equation}

 or equivalently:


\begin{equation}
\hat{y}_{n+1}= 
\begin{cases}
1,  & \text{if} \quad Pr(\hat{y}_{n+1}=1\vert \mathbf{x}_{n+1})= \sigma(\widehat{\boldsymbol{\theta}}^\top \mathbf{x}_{n+1}) >0.5 ,\\
0,  & \text{if} \quad Pr(\hat{y}_{n+1}=1\vert \mathbf{x}_{n+1})= \sigma(\widehat{\boldsymbol{\theta}}^\top \mathbf{x}_{n+1}) <0.5.
\end{cases} \nonumber
\end{equation}

**How to learn parameters, $\boldsymbol{\theta}=(\theta_0$, $\theta_1$, $\ldots$, $\theta_d)$ from training data $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$ to build up the decision boundary?**

## Conditional Maximum Likekihood Estimation for Logistic Regression

Since $Y_{i}$ is a `binary variable` and takes on either $0$ or $1$ values, we can assume that $Y_{i}$ given $\mathbf{x}_{i}$
follows a `Bernoulli distribution` as follows:

\begin{equation}
Y_{i}\vert \mathbf{x}_{i} \sim Bern(p_i), \nonumber
\end{equation}

where $p_i=Pr(y_i=1\vert \mathbf{x}_{i})=\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)$ is the probability that $Y_{i}$ given $\mathbf{x}_{i}$ belongs to 1.

Then the `probability mass function` of $Y_i$ can be written as:

\begin{eqnarray}
\begin{split}  
Pr(Y_{i}=y_i \vert \mathbf{x}_{i})&=p_{i}^{y_i}(1-p_{i})^{1-y_i} \\
&=\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)^{y_i}\big(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big)^{1-y_i} \quad \text{for} \quad y_i \in \{0,1\}.\nonumber
\end{split} 
\end{eqnarray}

## Likelihod Function

Then, the `conditional likelihood` function of the logistic regression for a **given training dataset** $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$ of size $n$ is given by:

\begin{eqnarray}
\begin{split}
\mathcal{L}(\boldsymbol{\theta}) &= \prod_{i=1}^{n} Pr(Y_{i}=y_i \vert \mathbf{x}_{i}) 
= \prod_{i=1}^{n} p_{i}^{y_i}(1-p_{i})^{1-y_i} \\
          &= \prod_{i=1}^{n} Pr(y_i=1\vert \mathbf{x}_{i})^{y_i}Pr(y_i=0\vert \mathbf{x}_{i})^{1-y_i}  \\ 
 &= \prod_{i=1}^{n} \exp\Big(y_i \log\big(Pr(y_i=1\vert \mathbf{x}_{i})\big) + (1-y_i)  \log\big(Pr(y_i=0\vert \mathbf{x}_{i})\big)\Big)  \\
 &=\exp\sum_{i=1}^{n}\Big(y_i \log\big(Pr(y_i=1\vert \mathbf{x}_{i})\big) + (1-y_i)  \log\big(Pr(y_i=0\vert \mathbf{x}_{i})\big)\Big).  \\
\nonumber          
\end{split} 
\end{eqnarray}


## Log-Likelihod Function

The `log-likelihood` function of the logistic regression for a **given training dataset** $\mathcal{D} = \{(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \ldots, (\mathbf{x}_{n}, y_{n})\}$ can be re-written as:


\begin{eqnarray}
\begin{split}
\ell(\boldsymbol{\theta})= \log(\mathcal{L}(\boldsymbol{\theta}))&=\sum_{i=1}^{n} \Big(y_i \log\big(Pr(y_i=1\vert \mathbf{x}_{i})\big) + (1-y_i)  \log\big(Pr(y_i=0\vert \mathbf{x}_{i})\big)\Big)\\
&=\sum_{i=1}^{n} \Big(y_i \log\big(\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big) + (1-y_i)  \log\big(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big)\Big)
.\nonumber  
\end{split}
\end{eqnarray}

The `log-likelihood` function measures how close the model’s predictions are to the actual y's.


## Negative Log-Likelihood: Binary Cross-Entropy

The `negative log-likelihood` function can be re-written as:


\begin{eqnarray}
\begin{split}
-\ell(\boldsymbol{\theta}) &=-\sum_{i=1}^{n} \Big(y_i \log\big(Pr(y_i=1\vert \mathbf{x}_{i})\big) + (1-y_i)  \log\big(Pr(y_i=0\vert \mathbf{x}_{i})\big)\Big)\\
&=-\sum_{i=1}^{n} \Big(y_i \log\big(\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big) + (1-y_i)  \log\big(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big)\Big)

.\nonumber  
\end{split}
\end{eqnarray}

In practice, we also call `negative log-likelihood` function as the `logistic loss function` or `binary cross-entropy` function.

## Maximum Likelihood Estimation

The `maximum likelihood estimation` of the log-likelihood function can be obtained as follows:


\begin{equation}
\hat{\boldsymbol{\theta}}_{MLE} = \underset{\boldsymbol{\theta}}{argmax} \quad \ell(\boldsymbol{\theta}), \nonumber
\end{equation}

which requires calculating the gradient vector:

\begin{equation}
\nabla_{\theta}\ell(\boldsymbol{\theta})=\frac{\partial  \ell(\boldsymbol{\theta})}{\partial\boldsymbol{\theta}}= \begin{bmatrix} 
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_0} \\
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_1} \\
\vdots \\
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_d}
\end{bmatrix}, \nonumber   
\end{equation}

and setting it to zero.


## Derivatives of the Log-Likelihood

The **individual components** of the gradient vector can be calculated as:

\begin{eqnarray}
\begin{split}
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_j}&=\frac{\partial \sum_{i=1}^{n} \Big(y_i \log\big(\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big) + (1-y_i)  \log\big(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big)\Big)}{\partial\theta_j} \\  
&=\sum_{i=1}^{n} \left( y_i \frac{1}{\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)} - (1-y_i) \frac{1}{1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)} \right) \frac{\partial}{\partial \theta_j} \sigma(\boldsymbol{\theta}^\top \textbf{x}_i)  \\
&=\sum_{i=1}^{n} \left( y_i \frac{1}{\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)} - (1-y_i) \frac{1}{1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)} \right)\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i))  \frac{\partial}{\partial \theta_j} \boldsymbol{\theta}^\top \textbf{x}_i \\
&=\sum_{i=1}^{n} \left( y_i \frac{1}{\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)} - (1-y_i) \frac{1}{1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)} \right)\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)) x_{ij} \\
&=\sum_{i=1}^{n} \left( y_i (1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)) - (1-y_i) \sigma(\boldsymbol{\theta}^\top \textbf{x}_i) \right)x_{ij} \\
&=\sum_{i=1}^{n} \left( y_i -\sigma(\boldsymbol{\theta}^\top \textbf{x}_i) \right) x_{ij}
\nonumber
\end{split}
\end{eqnarray}

for $j=0,1,\ldots, d$.


## Gradient of the Log-Likelihood

Then the **gradient vector** is equal to:

\begin{eqnarray}
\begin{split}
\nabla_{\theta}\ell(\boldsymbol{\theta}) &=\frac{\partial  \ell(\boldsymbol{\theta})}{\partial\boldsymbol{\theta}} = \begin{bmatrix} 
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_0} \\
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_1} \\
\vdots \\
\frac{\partial  \ell(\boldsymbol{\theta})}{\partial \theta_d}
\end{bmatrix} =
\textbf{x}^T \cdot \left( \textbf{y} -\sigma(\textbf{x} \boldsymbol{\theta}) \right).
\nonumber
\end{split}
\end{eqnarray}


## No-Closed form solution

The **coefficients** $\boldsymbol{\theta}$ enter into log-likelihood function in a **non-linear form**, unfortunately, calculating the gradient of the log-likelihood function with respect to the $\boldsymbol{\theta}$ and setting it to zero does not provide a **closed-form** solution for $\boldsymbol{\theta}$.

## Gradient Descent

The log-likelihood functions based on the statistical distributions belong to **exponential family** are `concave`. 
`Maximizing a concave function` (e.g., the log-likelihood function) is equivalent to `minimizing a convex` function (e.g., the binary cross-entropy function):


\begin{equation}
\hat{\boldsymbol{\theta}} = \underset{\boldsymbol{\theta}}{argmax} \quad \ell(\boldsymbol{\theta})= \underset{\boldsymbol{\theta}}{argmin} \quad -\ell(\boldsymbol{\theta}) \nonumber
\end{equation}

## Gradient Descent for Logistic Regression

We can **minimize** the binary cross-entropy through `gradient descent`.

**Algorithm**:

   -   Initialize $\boldsymbol{\theta}$.
   -   Repeat the following update rule until convergence:

   
 \begin{eqnarray}
    \begin{split}
    \boldsymbol{\theta}^{k+1} & = \boldsymbol{\theta}^{k} - \eta \left.\bigg(\nabla_{\theta}(-\ell(\boldsymbol{\theta}))\bigg) \right\vert_{k}, \\
      &= \boldsymbol{\theta}^{k} + \eta\left.\bigg(\nabla_{\theta}\ell(\boldsymbol{\theta})\bigg) \right\vert_{k}, 
    \\ \nonumber
    \end{split}
  \end{eqnarray}
 
for $\eta$ ($\eta>0$) is the `step-size` (`learning rate`), and $k$ is the `iteration number`.

Note that `gradient descent` is sensitive to the scale of the features, for that reason, numerical features should be pre-processed.

## Re-visit Diabetes Dataset

Diabetes data set contains details of a hospital's patients and the target variable is a binary variable reflecting the fact: whether the patient has diabetes (Yes:1) or not (No:0).

```python
#import the data set

import pandas as pd

diabetes_df = pd.read_csv("datasets/Diabetes.csv")

#there are some inconsistent values in the data set. I removed patients whose bmi and glucose are both zero
diabetes_df = diabetes_df[(diabetes_df.bmi!=0) & (diabetes_df.glucose!=0)]

#Prepare X and y
X = diabetes_df[["bmi","glucose"]]
y = diabetes_df[["diabetes"]]
```

```python
#first split the data as train and test. Use stratify so that the ratio of 0's to 1's in train data equals to that of in test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1773, stratify = y)
```

```python
#fit the logistic classifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

lr_pipe = make_pipeline(StandardScaler(), LogisticRegression(penalty=None))

lr_pipe.fit(X_train, y_train.diabetes)
```

```python
#print the learned categories
print(lr_pipe.named_steps['logisticregression'].classes_)
```

```python
#print intercept parameter estimates
print(lr_pipe.named_steps['logisticregression'].intercept_)
```

```python
#print slope parameter estimates
#print(lr_pipe.named_steps['logisticregression'].feature_names_in_)
print(lr_pipe.named_steps['logisticregression'].coef_)
```

## Uncertainity Estimates

There are two different function in scikit-learn that can be used to obtain uncertainity estimates from classifiers:


- `decision_function`: This value encodes how strongly the model believes a data point belong to the **positive class**, in this case class 1. Positive values indicate a preference for the positive class, and negative values indicate a preference for the "negative" class, in this case class 0. We can recover the prediction by looking at the sign of the decision function.

- `predict_proba`: The output of `predict_proba` is a probability for each class. The first entry in each row is the estimated probability of the first class, and the second entry is the estimated probability of the second class. The class with et least 50% certainity will be the one predicted.
 
Most, but not all, classifiers have at least one of them. 

```python
# Return the decision function score for each Xtest point
lr_pipe.decision_function(X_test.iloc[0:10])
```

```python
#Return the probability of being in a specific class for each Xtest point. 
lr_pipe.predict_proba(X_test.iloc[0:10])
```

```python
#Return the model predicted class labels
lr_pipe.predict(X_test.iloc[0:10])
```

```python
print("Accuracy on train data: %.2f %%" % (lr_pipe.score(X_train,y_train)*100))
```

```python
print("Accuracy on test data: %.2f %%" % (lr_pipe.score(X_test,y_test)*100))
```

## Penalized Logistic Regression

The `augmented objective function` of `penalized logistic regression with L2 norm`:

\begin{eqnarray}
\begin{split}
\underset{\boldsymbol{\theta}}{argmin} \quad -\sum_{i=1}^{n} \Big(y_i \log\big(\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big) + (1-y_i)  \log\big(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big)\Big) + \frac{1}{C} ||\boldsymbol{\theta}||_2^2.
\end{split}
\end{eqnarray}

The `augmented objective function` of `penalized logistic regression with L1 norm`:

\begin{eqnarray}
\begin{split}
\underset{\boldsymbol{\theta}}{argmin} \quad -\sum_{i=1}^{n} \Big(y_i \log\big(\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big) + (1-y_i)  \log\big(1-\sigma(\boldsymbol{\theta}^\top \textbf{x}_i)\big)\Big) + \frac{1}{C} ||\boldsymbol{\theta}||_1.
\end{split}
\end{eqnarray}

where C is the parameter **inverse to alpha**.

Let's fit penalized logistic regression with L1 and L2 norms to the diabetes data to find to the best model. This requires tuning the `hyper-parameter` C.

```python
from sklearn import set_config
set_config(transform_output="pandas")  

#first split the data as train and test. Use stratify so that the ratio of 0's to 1's in train data equals to that of in test data
from sklearn.model_selection import train_test_split

X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1773, stratify=y)
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

#The choice of the algorithm depends on the penalty chosen.
lr_pipe = make_pipeline(StandardScaler(), LogisticRegression(solver='liblinear'))

#Define parameter range
param_grid = [{'logisticregression__penalty' : ['l1', 'l2'],
               'logisticregression__C' : np.logspace(-4, 4, 20) #Return numbers spaced evenly on a log scale.
               }
              ]  
               
#The folds are made by preserving the percentage of samples for each class.
kcv_index = StratifiedKFold(n_splits=5, shuffle=True, random_state=1773) 

grid_search = GridSearchCV(estimator = lr_pipe, param_grid = param_grid, scoring='accuracy', cv = kcv_index, return_train_score=True) 
```

```python
grid_search.fit(X_train_val, y_train_val.diabetes)
```

```python
#you can investigate validation accuracy in each fold for a given C
#here test score = validation score
#pd.DataFrame(grid_search.cv_results_)
```

```python
print("Optimum Params:", grid_search.best_params_)
print("CV Validation Accuracy: %.2f %%" % (grid_search.best_score_*100))
```

```python
grid_search.best_estimator_
```

```python
#Evaluate performance of best logistic classifier on train+val
print("Accuracy on train+val data: %.2f %%" % (grid_search.score(X_train_val, y_train_val)*100))
```

```python
#Evaluate final performance of best logistic classifier on test
print("Accuracy on test data: %.2f %%" % (grid_search.score(X_test, y_test)*100))
```

## Decision Boundaries 

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [18, 8]

from sklearn.inspection import DecisionBoundaryDisplay

fig, ax = plt.subplots(1,2)

#get transformed matrices
X_train_val_scaled = grid_search.best_estimator_[0].fit_transform(X_train_val)
X_test_scaled = grid_search.best_estimator_[0].transform(X_test)
        
#left plot: plot the training+val data on the decisionboundary 
db = DecisionBoundaryDisplay.from_estimator(grid_search.best_estimator_[1], X_train_val_scaled, response_method="predict", 
                                            xlabel="BMI", ylabel="Glucose", ax = ax[0], cmap="Reds", alpha=0.9) 
                
db.ax_.scatter(X_train_val_scaled.bmi, X_train_val_scaled.glucose, c=y_train_val.diabetes, edgecolor="black")
db.ax_.set_title('Accuracy of Logistic Classifier with L1 norm on train+val data: %.2f%%' % (grid_search.score(X_train_val, y_train_val)*100))


#right plot: plot the test data on the decisionboundary 
db = DecisionBoundaryDisplay.from_estimator(grid_search.best_estimator_[1], X_train_val_scaled, response_method="predict", 
                                            xlabel="BMI", ylabel="Glucose", ax = ax[1], cmap="Reds", alpha=0.9) 
db.ax_.scatter(X_test_scaled.bmi, X_test_scaled.glucose, c=y_test.diabetes, edgecolor="black")
db.ax_.set_title('Accuracy of Logistic Classifier with L1 norm on test data: %.2f%%' % (grid_search.score(X_test, y_test)*100))

plt.show();
```
